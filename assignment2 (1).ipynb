```python
#Finding average
Average marks are calculated for each group to compare performance across categories.
```

```python
import pandas as pd
df = pd.read_csv("student_bias_dataset.csv")
avg_marks = df.groupby('Gender')['Marks'].mean()
print(avg_marks)
```

```python
# Visualization
Bar charts and box plots are used to visually understand distribution and variation.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("student_bias_dataset.csv")
group = df['Gender'].value_counts()
print(group)
group.plot(kind ='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender ")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("student_bias_dataset.csv")
group = df['Gender'].value_counts()
print(group)
df.boxplot(column ='Marks', by = 'Gender')
plt.title("Marks Distribution By Gender")
plt.suptitle("")
plt.xlabel("Gender")
plt.ylabel("Marks")
plt.show()
print(df[['Study_Hours', 'Marks']].corr())
```

```python
## Key Insights

- The dataset shows an imbalance in gender distribution, where one group has more representation than the other.
- There is a noticeable difference in average marks between categories, indicating variation in performance.
- Some categories are underrepresented, which may lead to biased conclusions.
- The dataset does not fully represent all groups equally, which can affect fairness in analysis.
```
